```r
data("UScereal")
df <- UScereal
head(df)
```

```r
df1 <- df[,c(-1,-2,-9,-11)]
head(df1)
```

```r
calories<-df$calories
protein<-df$protein
fat<-df$fat
sodium<-df$sodium
fibre<-df$fibre
carbo<-df$carbo
sugars<-df$sugars
potassium<-df$potassium
```

```r
model<-lm(calories~protein+fat+sodium+fibre+carbo+sugars+potassium)
summary(model)
```

```r
model2<-lm(calories~protein+fat+fibre+carbo+sugars+potassium)
summary(model2)
```

```r
library(psych)
```

```r
X <- data.matrix(df1)
pairs.panels(X,smooth=F)
```

```r
pca<-princomp(X,cor=T)
summary(pca)
```

```r
plot(1:7,pca$sd^2,pch=19,xlab="Scaled PCA",ylab="Eigen-Values",main="Scree Plot")
lines(c(1:7),pca$sd^2)
```

```r
install.packages("scatterplot3d", dependencies=TRUE, repos='http://cran.rstudio.com/')
library(scatterplot3d)
```

```r
scatterplot3d(pca$scores[,1],pca$scores[,2],pca$scores[,3],pch = 19, color = "steelblue")
```
