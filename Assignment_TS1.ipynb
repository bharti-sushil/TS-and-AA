### Now Checking for Most Appropriate ARIMA Model

```r
arma_aics<-function(x,P,d,Q)
{
  aics<-matrix(nrow=P+1,ncol=Q+1)
  for(p in 0:P)
    for(q in 0:Q)
    {
      mdl<-arima(x,order=c(p,d,q),method = "ML")
      if( mdl$code==0 ) aics[p+1,q+1]<-mdl$aic
    }
  return(aics)
}
```

### Akaike Information Criterion to find best suited model 

```r
aic10<-arma_aics(lPrice,10,1,10)
aic10
```

### Sorting AIC in order to understand their relative significance

```r
normtest<-function(x)
{
  library(nortest)
  s<-shapiro.test(x)
  ad<-ad.test(x)
  cvm<-cvm.test(x)
  ll<-lillie.test(x)
  sf<-sf.test(x)
  df<-data.frame(Method=c(s$method, ad$method, cvm$method, ll$method, sf$method),
                 P.Value=c(s$p.value, ad$p.value, cvm$p.value, ll$p.value, sf$p.value))
  df
}
```

## IRF for Selected Model i.e. model3_3

### Creating IRF plot function

```r
irfplot<-function (irf, s)
{
  n <- length(irf)
  plot(c(0,n+1), range(c(irf,1)), type = "n",xlab = "Time", ylab = "IRF", main = s)
  lines(c(0,n+1),c(0,0))
  lines(c(0,0),c(0,1))
  for (i in 1:n)
    lines(c(i,i), c(0, irf[i]))
}
```

```r
# IRF of model model3_3
par(mfrow=c(1,2))
psi<-ARMAtoMA(ar = c(0.5653, -0.8193,0.5861),ma=c(-0.4649, 0.8527, -0.5535),lag.max=30)
irfplot(psi,"ERR IRF")
irfplot(cumsum(psi)+1,"Log-ERR IRF")
```

```r
# Autocorrelation
gamma0<-(1+sum(psi^2))*7.549e-05
gamma0
```

```r
psi100<-ARMAtoMA(ar = c(0.5653, -0.8193,0.5861),ma=c(-0.4649, 0.8527, -0.5535),lag.max=100)
gamma0<-(1+sum(psi100^2))*7.549e-05
gamma0
```
